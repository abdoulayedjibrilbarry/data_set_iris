```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np

# Charger le jeu de données Iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data.head()
```

```python
mean_sepal_length = data['sepal length (cm)'].mean()
median_sepal_length = data['sepal length (cm)'].median()
print(f"Min sepal lengt: {round(mean_sepal_length, 3)} median sepal length: {median_sepal_length} ")
```

```python
percentile_75 = data.groupby('species')['petal length (cm)'].quantile(0.75)
print(percentile_75)
```

```python
data['sepal_area'] = data['petal length (cm)'] * data['sepal width (cm)']
data.head()
```

```python
def remove_outliers(group):
    std = group['petal length (cm)'].std()
    return group[group['petal length (cm)'] <= 2 * std]

iris_no_outliers = data.groupby('species', group_keys=False).apply(remove_outliers)
print(f"4️⃣ Nombre de lignes après suppression des outliers : {len(iris_no_outliers)} (au lieu de {len(data)})\n")
```

```python


# Créer le scaler
scaler = MinMaxScaler()

# Normaliser toutes les colonnes numériques (sauf 'species')
data_scaled = data.copy()
data_scaled[data.columns[:-1]] = scaler.fit_transform(data[data.columns[:-1]])
print(data_scaled.head())
```

```python
subset = data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)']]
top3 = subset.value_counts().head(3)
print(top3)
```

```python
data['species'] = [iris.target_names[i] for i in iris.target]
idx = data.groupby('species') ['sepal width (cm)'].idxmax()
print(idx)
```

```python
import numpy as np

np.random.seed(0)
random_indices = np.random.choice(data.index , size=5, replace=False)
data.loc[random_indices, 'petal width (cm)'] *= -1

print(data.loc[random_indices, ['petal width (cm)']])
```

```python
mean_non_negative = data.loc[data['petal width (cm)'] >= 0, 'petal width (cm)'].mean()
print(mean_non_negative)
```

```python
data.loc[data['petal width (cm)'] < 0, 'petal width (cm)'] = mean_non_negative
```
